```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

kannada_texts = [
    "ಕರ್ನಾಟಕದ ಸಂಸ್ಕೃತಿ ಬಹಳ ಹಳೆಯದು ಮತ್ತು ವೈವಿಧ್ಯಮಯವಾಗಿದೆ.",
    "ವಿಶ್ವದ ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ಭಾರತದ ಪ್ರಾಚೀನ ಶಾಸ್ತ್ರಗಳು ಪ್ರಮುಖ ಸ್ಥಾನವನ್ನು ಹೊಂದಿವೆ.",
    "ಬೆಂಗಳೂರನ್ನು ಭಾರತದ ಸಿಲಿಕಾನ್ ವ್ಯಾಲಿ ಎಂದು ಕರೆಯಲಾಗುತ್ತದೆ.",
    "ಅನೇಕ ಕವಿ ಕೃತಿಗಳು ಕನ್ನಡ ಸಾಹಿತ್ಯದಲ್ಲಿ ಮಹತ್ವದ ಸ್ಥಾನವನ್ನು ಪಡೆದಿವೆ.",
    "ಭಾರತದ ಸ್ವಾತಂತ್ರ್ಯ ಹೋರಾಟದಲ್ಲಿ ಮಹಾತ್ಮ ಗಾಂಧಿಯವರು ಪ್ರಮುಖ ಪಾತ್ರವಹಿಸಿದರು.",
    "ಕನ್ನಡ ಭಾಷೆಯ ವ್ಯಾಕರಣವನ್ನು ಬಹಳ ಸರಳವಾಗಿ ಕಲಿಯಬಹುದು.",
    "ಐತಿಹಾಸಿಕ ತಜ್ಞರು ಹಲವಾರು ಶಾಸನಗಳನ್ನು ಕನ್ನಡದಲ್ಲಿ ಪತ್ತೆಹಚ್ಚಿದ್ದಾರೆ.",
    "ವೃತ್ತಿಪರ ಅಭ್ಯಾಸಗಳಲ್ಲಿ ಸಂಶೋಧನೆ ಮತ್ತು ಬೆಳವಣಿಗೆಯ ಅವಶ್ಯಕತೆ ಹೆಚ್ಚಾಗಿದೆ.",
    "ನಮ್ಮ ಜಲವಾಯು ಪರಿಸರ ಸಂರಕ್ಷಣೆಗಾಗಿ ಸಂವೇದನೆ ಮೂಡಿಸಲು ನವೀನ ತಂತ್ರಜ್ಞಾನಗಳು ಅಗತ್ಯವಿವೆ.",
    "ಆಧುನಿಕ ಶಿಕ್ಷಣ ವಿಧಾನಗಳು ಕೌಶಲ್ಯಾಭಿವೃದ್ಧಿಗೆ ನೆರವಾಗುತ್ತವೆ.",
    "ಭಾರತೀಯ ಚಲನಚಿತ್ರ ಉದ್ಯಮದಲ್ಲಿ ಕನ್ನಡ ಚಿತ್ರಗಳು ಪ್ರಭಾವಶಾಲಿಯಾಗಿವೆ.",
    "ತಂತ್ರಜ್ಞಾನದ ಸಹಾಯದಿಂದ ಆರ್ಥಿಕತೆಯಲ್ಲಿ ಶ್ರೇಷ್ಟತೆ ಸಾಧಿಸಬಹುದು.",
    "ಪ್ರತಿ ಕನ್ನಡಿಗನೂ ತನ್ನ ಭಾಷೆಯನ್ನು ಗೌರವದಿಂದ ನೋಡಬೇಕು.",
    "ಅರ್ಥಶಾಸ್ತ್ರದಲ್ಲಿ ಕನ್ನಡದ ಪ್ರಮುಖ ಗ್ರಂಥಗಳು ಲಭ್ಯವಿವೆ.",
    "ಪ್ರಕೃತಿಯ ವೈವಿಧ್ಯತೆಯನ್ನು ನಮ್ಮ ಸಂಸ್ಕೃತಿಯಲ್ಲಿ ವಿಶೇಷವಾಗಿ ಗಮನಿಸಲಾಗಿದೆ.",
    "ಭಾರತೀಯ ಗಣಿತಜ್ಞರು ಅನೇಕ ಗಣಿತ ನಿಯಮಗಳನ್ನು ಕಂಡುಹಿಡಿದಿದ್ದಾರೆ.",
    "ಶಿಕ್ಷಣ ನಮ್ಮ ದೇಶದ ಅಭಿವೃದ್ಧಿಗೆ ಮುಖ್ಯವಾದ ಸಾಧನವಾಗಿದೆ.",
    "ಕೃಷಿ ಕ್ಷೇತ್ರದಲ್ಲಿ ನಾವೀನ್ಯತೆ ತರಲು ಇಂಧನ ಸಂಪತ್ತು ಕಾಪಾಡಬೇಕಾಗಿದೆ.",
    "ಸಂಪ್ರದಾಯಗಳು ಮತ್ತು ಆಧುನಿಕತೆಯ ಸಂಯೋಜನೆ ನಮ್ಮ ಸಮಾಜವನ್ನು ಮುಂದೆ ಕರೆದೊಯ್ಯುತ್ತದೆ.",
    "ಇಂಟರ್ನೆಟ್ ಬಳಸುವವರ ಸಂಖ್ಯೆ ದಿನದಿಂದ ದಿನಕ್ಕೆ ಹೆಚ್ಚುತ್ತಿದೆ.",
    "ಸಾಹಿತ್ಯ ಮತ್ತು ಕಲೆ ನಮ್ಮ ಶ್ರೇಷ್ಠತೆಯನ್ನು ಪ್ರತಿಬಿಂಬಿಸುತ್ತವೆ.",
    "ಐಕಾನ್ ಆಗಿರುವ ಆಧುನಿಕ ಕನ್ನಡ ಗೀತೆಗಳು ಜನಪ್ರಿಯವಾಗಿವೆ.",
    "ಆಧುನಿಕ ಆಯುರ್ವೇದ ಚಿಕಿತ್ಸೆಯಲ್ಲಿ ನವೀನ ಬದಲಾವಣೆಗಳುಂಟಾಗಿವೆ.",
    "ಅವರ ತತ್ವಶಾಸ್ತ್ರಗಳು ಇಂದಿಗೂ ಪ್ರಸ್ತುತವಾಗಿವೆ.",
    "ಭಾರತದ ಪೌರಾಣಿಕ ಕಥೆಗಳು ಯುವಪೀಳಿಗೆಗೆ ಪಾಠಗಳನ್ನು ನೀಡುತ್ತವೆ.",
    "ಸಮಾಜದಲ್ಲಿ ಬದಲಾಗುತ್ತಿರುವ ಪ್ರಗತಿಶೀಲ ಚಿಂತನೆಗಳು ತಂತ್ರಜ್ಞಾನವನ್ನು ಪ್ರೋತ್ಸಾಹಿಸುತ್ತವೆ.",
    "ಸಂಗೀತ ನಮ್ಮ ಮನಸ್ಸುಗಳನ್ನು ಸಮಾಧಾನಕ್ಕೆ ತರಲು ಸಹಕಾರಿಯಾಗಿದೆ.",
    "ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ಹಿಂದಿನ ಸಂಸ್ಕೃತಿಯ ಅಧ್ಯಯನಕ್ಕೆ ಪ್ರಾಮುಖ್ಯತೆ ನೀಡಲಾಗುತ್ತಿದೆ.",
    "ಪ್ರಕೃತಿ ಸಂರಕ್ಷಣೆಯಲ್ಲಿ ಸಜೀವ ಸಂಕುಲದ ಪಾತ್ರ ಮಹತ್ವದಾಗಿದೆ.",
    "ಆರ್ಥಿಕ ಹೋರಾಟಗಳು ಸಾಮಾಜಿಕ ಬದಲಾವಣೆಗೆ ಕಾರಣವಾಗುತ್ತವೆ."
]

for i in range(70):
    kannada_texts.append(f"ಸಂಪಾದನೆ {i+1}: ಕನ್ನಡದಲ್ಲಿ ಇನ್ನಷ್ಟು ವಿಷಯಗಳನ್ನು ಸೇರಿಸಲಾಗಿದೆ.")

query_texts = kannada_texts[:20]
doc_texts = kannada_texts[20:]
models = {
    "LaBSE": SentenceTransformer('sentence-transformers/LaBSE'),
    "DistilUSE": SentenceTransformer("sentence-transformers/distiluse-base-multilingual-cased-v2"),
    "STSB-XLM-R": SentenceTransformer("sentence-transformers/stsb-xlm-r-multilingual"),
    "Paraphrase-MPNet": SentenceTransformer("sentence-transformers/paraphrase-multilingual-mpnet-base-v2"),
    "All-MPNet": SentenceTransformer("sentence-transformers/all-mpnet-base-v2"),
    "Kannada-Similarity-SBERT": SentenceTransformer("l3cube-pune/kannada-sentence-similarity-sbert"),
    "Indic-SBERT-NLI": SentenceTransformer("l3cube-pune/indic-sentence-bert-nli"),
    "MiniLM-Multilingual": SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"),
    "DistilRoBERTa-Multilingual": SentenceTransformer("sentence-transformers/paraphrase-distilroberta-base-v1")
}

def precision_at_k(rankings, ground_truth, k=5):
    precisions = []
    for i, ranking in enumerate(rankings):
        relevant = set(ground_truth[i])
        retrieved = set(ranking[:k])
        precisions.append(len(relevant & retrieved) / k)
    return sum(precisions) / len(precisions)

def recall_at_k(rankings, ground_truth, k=5):
    recalls = []
    for i, ranking in enumerate(rankings):
        relevant = set(ground_truth[i])
        retrieved = set(ranking[:k])
        recalls.append(len(relevant & retrieved) / len(relevant))
    return sum(recalls) / len(recalls)

def ndcg_at_k(rankings, ground_truth, k=5):
    def dcg(relevances, k):
        return sum(rel / np.log2(idx + 2) for idx, rel in enumerate(relevances[:k]))
    ndcg_scores = []
    for i, ranking in enumerate(rankings):
        relevances = [1 if doc in ground_truth[i] else 0 for doc in ranking]
        dcg_score = dcg(relevances, k)
        idcg_score = dcg(sorted(relevances, reverse=True), k)
        ndcg_scores.append(dcg_score / idcg_score if idcg_score > 0 else 0)
    return sum(ndcg_scores) / len(ndcg_scores)

results = {}
ground_truth = [[j] for j in range(len(query_texts))]

for model_name, model in models.items():
    query_embeddings = model.encode(query_texts, convert_to_tensor=True)
    doc_embeddings = model.encode(doc_texts, convert_to_tensor=True)
    similarities = cosine_similarity(query_embeddings.cpu().numpy(), doc_embeddings.cpu().numpy())
    rankings = similarities.argsort(axis=1)[:, ::-1]
    results[model_name] = {
        "Precision@5": precision_at_k(rankings, ground_truth, k=5),
        "Recall@5": recall_at_k(rankings, ground_truth, k=5),
        "NDCG@5": ndcg_at_k(rankings, ground_truth, k=5)
    }

for model_name, metrics in results.items():
    print(f"{model_name}: {metrics}")
```
